#Google Colaboratory - Importing and exporting data into runtime

* Best option is to pre-process the data locally, using multicore CPUs and software, then sync the results with Google Drive. This pre-processed data can then be efficiently imported into Google Colab for GPU processing
* Note that data uploaded into a runtime will be automatically deleted when the session is terminated

## Define helper functions for notebook

```
# function to display an rgb image

def show_image_rgb(image, title):
    
    import matplotlib.pyplot as plt
    
    plt.imshow(image)
    plt.title(title)    
    plt.xticks([])
    plt.yticks([])
    plt.show()
    plt.close()
    
    return
```

##Method 1 : Upload file from local machine using side panel

* Import data file using side panel [Files] > [upload] function

```
# read uploaded image file   
  
from skimage.io import imread

image_name = 'perceptron.jpg'
image = imread(image_name)

# display image

show_image_rgb(image, image_name)
```

## Method 2 : Upload by mounting Google Drive to the runtime

* Mount Google Drive to virtual machine using an authorization code
* The root directory for the virtual machine is: /content
* After mounting, the Google Drive structure appears in the left side panel, might need to refresh
* Can copy the path of a file by right clicking on the file in the left side panel
* Changes to the local Google Drive are synced to the virtual machine, and vica-versa; might be better to turn off Google Drive sync to local machines until session is finished to prevent massive internet traffic flow!
*  If the data is already stored on Google Drive in the cloud, the data i/o speeds into the runtime are faster than if uploaded from the local machine
* Python in Linux uses uses '/' in filepaths, whilst windows uses '\\\\'
* Can now use standard i/o Python library commands to import and export any data types: images, numpy, text, csv, and HDF5


```
# mount google drive to virtual machine using an authorization code

from google.colab import drive

drive.mount('/content/drive')
```

```
# read uploaded image file   
  
from skimage.io import imread

image_name = '/content/drive/My Drive/Colab Notebooks/DLOI/2019-2020/lectures/surface wave.png'  # full path
image = imread(image_name)

# display image

show_image_rgb(image, image_name)
```
