# Importing libraries

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
import skimage
from skimage import transform
from tensorflow import keras
```

# Loading MNIST

```
ix=28*28 # Number of features
digits=[0,1] # Digits to classify

# Loading data
mnist=keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

# Number of samples of each number

```
print('----')
print('Training samples: {:}'.format(len(train_images)))
for i in range(10):
  a=np.where(train_labels==i)[0]
  print('There are {:} images of {:}'.format(len(a),i))
print('----')
print('Training samples: {:}'.format(len(test_images)))
for i in range(10):
  a=np.where(test_labels==i)[0]
  print('There are {:} images of {:}'.format(len(a),i))
```

# Plotting 5 different versions of the same number

```
digit=9
a=np.where(train_labels==digit)[0]
for i in range(5):
  plt.subplot(2,3,i+1)
  plt.imshow(train_images[a[i]],cmap='viridis')
  plt.grid(False);plt.axis('off');
```

# Sequence of character images from 0 to 9

```
for i in range(10):
  plt.subplot(2,5,i+1)
  a=np.where(train_labels==i)[0]
  plt.imshow(train_images[a[0]],cmap='viridis')
  plt.grid(False);plt.axis('off');
```
